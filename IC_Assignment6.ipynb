```python
#import numpy
import numpy as np
```

```python
#Create an algorithm that calculates the population of 3 ages of a beetle population and the total population over 50 years
#The initial First year population is 500 and the initial second year population is 500
#On average 1 beetle produces 1 offspring in second year and 5 offspring in third year
#The first year beetles survive 30% to second year and 50% of second year beetles survive to third year

#Create a definition to calculate and return the population of each age group and the total population for each year with an initial population of 500 first year and 500 second year beetles
def beetle_population(years, initial_first_year, initial_second_year):
    #set explanation of parameters
    """
    Parameters:
    years : integer
         number of years to calculate the population for
    initial_first_year: integer
        initial population of first year beetles
    initial_second_year: integer
        initial population of second year beetles
    Returns:
    first_year_population (list): population of first year beetles for each year
    second_year_population (list): population of second year beetles for each year
    third_year_population (list): population of third year beetles for each year
    total_population (list): total population of beetles for each year
    """
    #Complete the algorithm using matrix multiplication
    #Create a matrix to represent the transition of beetle populations from one year to the next
    A = np.array([[0, 1, 5], [0.3, 0, 0], [0, 0.5, 0]])
    #initialize the population vector with initial populations
    initial_population_vector = np.array([initial_first_year, initial_second_year, 0])
    #Create a list to store the total population for each year
    first_year_population = []
    second_year_population = []
    third_year_population = []
    total_population = []
    #append the current populations to the lists
    first_year_population.append(initial_population_vector[0])
    second_year_population.append(initial_population_vector[1])
    third_year_population.append(initial_population_vector[2])
    total_population.append(sum(initial_population_vector))
    #Loop through the number of years to calculate the population for each year
    n_t = initial_population_vector
    for year in range(years):
        #Calculate the population for the next year using matrix multiplication
        #loop through the number of years to calculate the populations 
        n_t = np.dot(A, n_t)
        #append the current populations to the lists
        first_year_population.append (n_t[0])
        second_year_population.append (n_t[1])
        third_year_population.append (n_t[2])
        total_population.append (sum(n_t))
    #Return the total population for each year
    return first_year_population, second_year_population, third_year_population, total_population
```

```python
# call the function to calculate the population for 150 years with initial populations of 500 first year and 500 second year beetles

first_year_population, second_year_population, third_year_population, total_population = beetle_population(150, 500, 500)
```

```python
# plot the total populations of first year, second year, and third year beetles over time
import matplotlib.pyplot as plt
plt.plot(first_year_population, label='First Year Beetles')
plt.plot(second_year_population, label='Second Year Beetles')
plt.plot(third_year_population, label='Third Year Beetles')
plt.plot(total_population, label='Total Beetle Population', linestyle='--', color='black')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Beetle Population Over Time')
plt.legend()
plt.show()
```

```python
#Create a definition to calculate and return the population of each age group and the total population for each year with an initial population of 500 first year and 500 second year beetles
def beetle_populationwithtrap(years, initial_first_year, initial_second_year):
    #set explanation of parameters
    """
    Parameters:
    years : integer
         number of years to calculate the population for
    initial_first_year: integer
        initial population of first year beetles
    initial_second_year: integer
        initial population of second year beetles
    Returns:
    first_year_population (list): population of first year beetles for each year
    second_year_population (list): population of second year beetles for each year
    third_year_population (list): population of third year beetles for each year
    total_population (list): total population of beetles for each year
    """
    #Complete the algorithm using matrix multiplication
    #Create a matrix to represent the transition of beetle populations from one year to the next
    # Update the transition matrix to reflect 5% removal from each age group due to traps
    A = np.array([
        [0, 0.95, 4.75],    # 5% reduction in second and third year beetles producing offspring
        [0.285, 0, 0],      # 5% reduction in first year beetles surviving to second year
        [0, 0.475, 0]       # 5% reduction in second year beetles surviving to third year
    ])
    #initialize the population vector with initial populations
    initial_population_vector = np.array([initial_first_year, initial_second_year, 0])
    #Create a list to store the total population for each year
    first_year_population = []
    second_year_population = []
    third_year_population = []
    total_population = []
    #append the current populations to the lists
    first_year_population.append(initial_population_vector[0])
    second_year_population.append(initial_population_vector[1])
    third_year_population.append(initial_population_vector[2])
    total_population.append(sum(initial_population_vector))
    #Loop through the number of years to calculate the population for each year
    n_t = initial_population_vector
    for year in range(years):
        #Calculate the population for the next year using matrix multiplication
        #loop through the number of years to calculate the populations 
        n_t = np.dot(A, n_t)
        #append the current populations to the lists
        first_year_population.append (n_t[0])
        second_year_population.append (n_t[1])
        third_year_population.append (n_t[2])
        total_population.append (sum(n_t))
    #Return the total population for each year
    return first_year_population, second_year_population, third_year_population, total_population
```

```python
first_year_population, second_year_population, third_year_population, total_population = beetle_populationwithtrap(150, 500, 500)
```

```python
# plot the first year, second year, and third year beetle populations over time with traps
import matplotlib.pyplot as plt
plt.plot(first_year_population, label='First Year Beetles with Traps')
plt.plot(second_year_population, label='Second Year Beetles with Traps')
plt.plot(third_year_population, label='Third Year Beetles with Traps')
plt.plot(total_population, label='Total Beetle Population with Traps', linestyle='--', color='black')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Beetle Population Over Time with Traps')
plt.legend()
plt.show()
```
