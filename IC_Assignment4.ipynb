```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
def NandM_population(N_0, M_0, time, K_N, K_M, d_N, d_M, S):
    #growth model
    # save N_values into a list
    N_values = [N_0]
    M_values = [M_0]
    #save initial population
    current_N = N_0
    current_M = M_0
    #set current population size
    # N_0= 5
    # M_0= 12
    # S=1
    # time=30
    # K_N= 400
    # K_M= 1600
    # d_N= 0.6
    # d_M= 0.6

    for _ in range(time):
        current_N= current_N + S * current_N * (1 - (current_N + d_N * current_M) / K_N)
        #round to three decimal places
        current_N= round(current_N, 3)
        N_values.append(current_N)

        current_M= current_M + S * current_M * (1 - (current_M + d_M * current_N) / K_M)
        #round to three decimal places
        current_M= round(current_M, 3)
        M_values.append(current_M)

    return N_values, M_values
```

```python
# print results
N_values, M_values = NandM_population(N_0=5, M_0=12, time=30, K_N=400, K_M=1600, d_N=0.4, d_M=0.6, S=1)
print("N values:", N_values)
print("M values:", M_values)
```

```python
# plot results
plt.plot(N_values, label='Species N', color='blue')
plt.plot(M_values, label='Species M', color='orange')
#add labels and title
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.title('Lotka-Volterra Competition: Population Over Time')
plt.grid(True)
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
#label the lines as Species N and Species M
```
