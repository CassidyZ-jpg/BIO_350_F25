```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
def NandM_population(N_0, M_0, K_N, K_M, S, d_N, d_M, time):
    # Exponential growth model
    # Save N_values into a list
    N_values = [N_0]
    M_values = [M_0]
    # set current population to initial population
    current_N = N_0
    current_M = M_0
    # Loop through time steps
    for _ in range(time):
        current_N = current_N + S * current_N * (1 - (current_N / K_N) - (current_M / K_M)) - d_N * current_N
        current_M = current_M + S * current_M * (1 - (current_M / K_M) - (current_N / K_N)) - d_M * current_M
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        current_M = round(current_M, 3)
        # Append to list
        N_values.append(current_N)
        M_values.append(current_M)
    return N_values, M_values
# Parameters
S = 1 # intrinsic growth rate
d_N = 0.4 # death rate of species N
d_M = 0.6 # death rate of species M
K_N = 400 # carrying capacity of species N
K_M = 1600 # carrying capacity of species M
N_0 = 12 # initial population of species N
M_0 = 5 # initial population of species M
time = 38 # time in years
```

```python
# run the model with time = 38, K_N = 400, S=1, K_M = 1600, d_N = 0.4, d_M = 0.6, N_0 = 12, M_0 = 5
N_values, M_values = NandM_population(N_0=12, M_0=5, time=38, K_N=400, K_M=1600, d_N=0.4, d_M=0.6, S=1)
#print results
print(N_values)
print(M_values)
```

```python
#plot results
import matplotlib.pyplot as plt

plt.plot(N_values, label='Species N')
plt.plot(M_values, label='Species M')
plt.xlabel('Time (years)')
plt.ylabel('Population Size')
plt.title('Population Dynamics of Species N and M')
# make the x and y axes start at 0
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
plt.show()
```

```python
#when does species M's population reach 0?
for year, population in enumerate(M_values):
    if population <= 0:
        print(f"Species M's population reaches 0 at year {year}.")
        break
    if year == len(M_values) - 1:
        print("Species M's population does not reach 0.")
```
