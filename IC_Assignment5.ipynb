```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# SIR model parameters
#import the variables given in the problem
beta = 0.001  # Infection rate
S0 = 999       # Initial susceptible population
I0 = 1        # Initial infected population
R0 = 0         # Initial removed population
d = 0.5     # death
dI_t = d * I0 # the number of infected foxes who die each day
t = 30       # Time in days
```

```python
# create a function that uses a for loop to calculate the number of susceptible, infected, and removed individuals over time
def sir_model(S0, I0, R0, beta, d, t):
    S = [S0]
    I = [I0]
    R = [R0]
    for day in range(1, t + 1):
        new_infections = beta * S[-1] * I[-1]
        new_deaths = d * I[-1]
        S.append(S[-1] - new_infections)
        I.append(I[-1] + new_infections - new_deaths)
        R.append(R[-1] + new_deaths)
    return S, I, R
```

```python
# print results of S, I, R
S, I, R = sir_model(999, 1, 0, 0.001, 0.5, 30)
for day in range(t + 1):
    print(f"Day {day}: S={S[day]:.2f}, I={I[day]:.2f}, R={R[day]:.2f}")
```

```python
#create a plot showing the number of susceptible, infected, and removed individuals over time
plt.figure(figsize=(10, 6))
plt.plot(S, label='Susceptible', color='blue')
plt.plot(I, label='Infected', color='red')
plt.plot(R, label='Removed', color='green')
plt.xlabel('Days')
plt.ylabel('Number of Individuals')
plt.title('SIR Model Over Time')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
plt.grid()
plt.show()
```
