```python
import matplotlib.pyplot as plt
```

```python
#logistic growth function
def logistic_growth(t, P0, r, K):
    # Save N_values into a list
    N_values = [P0]
    # Save initial population
    current_N = P0
    # Iterate for the specified time (t=0)
    for _ in range(t):
        # Update population using logistic growth formula
        #Nt+1 = Nt +(bNt-dNt)
        current_N = current_N + r * current_N * (1 - current_N / K)
        # Round population to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
        # Append new population to the list
    return N_values
# Parameters
t= 35 # time in days
P0 = 12  # initial population
r = 0.6  # growth rate
K = 400  # carrying capacity
```

```python
result= logistic_growth(35, 12, 0.6, 400)
print(result)
```

```python
# Plot the results with days on the x-axis and population on the y-axis
days = list(range(36))  # 0 to 35 days
plt.plot(days, result, marker='o')   
# Display the results of the exponential growth equation on a plot
plt.xlabel('Days')
# Label x-axis 'Days'
plt.ylabel(' Plant Population')
# Label y-axis 'Plant Population'
plt.title('Plant Population Growth Over Time')
# Title the plot 'Plant Population Growth Over Time'
#make the x-and y-axis both start at 0
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.grid(True)  
# Add grid lines to the plot
plt.show()
# Show the plot
```

```python
#logistic growth function
def logistic_growth(t, P0, s, d, K):
    # Save N_values into a list
    N_values = [P0]
    # Save initial population
    current_N = P0
    # Iterate for the specified time (t=0)
    for _ in range(t):
        # Update population using logistic growth formula
        #Nt+1 = Nt +(bNt-dNt)
        current_N = current_N + s * current_N * (1 - current_N / K) - d * current_N
        # Round population to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
        # Append new population to the list
    return N_values
# Parameters
t= 35 # time in days
P0 = 12  # initial population
s = 1 # number of seeds produced per plant
d= 0.4 # death rate
K = 400  # carrying capacity
```

```python
result= logistic_growth(35, 12, 1, 0.4, 400)
print(result)
```

```python
#set max y-axis limit to carrying capacity
plt.ylim(0, K)
# Plot the results with days on the x-axis and population on the y-axis
days = list(range(36))  # 0 to 35 days
plt.plot(days, result, marker='o')   
# Display the results of the exponential growth equation on a plot
plt.xlabel('Days')
# Label x-axis 'Days'
plt.ylabel(' Plant Population')
# Label y-axis 'Plant Population'
plt.title('Self-Shading Competition Model of Plant Population Growth Over Time')
# Title the plot 'Plant Population Growth Over Time'
#make the x-and y-axis both start at 0
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.grid(True)  
# Add grid lines to the plot
plt.show()
# Show the plot
```
