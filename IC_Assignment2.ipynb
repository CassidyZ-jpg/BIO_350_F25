- Calculating how quickly duckweed will cover a pond


```python
import matplotlib.pyplot as plt
# Install matplotlib if you don't have it: pip install matplotlib


```

```python
def N_exponential(N_0, p, time):
    # Exponential growth model
    # Save N_values into a list
    N_values = [N_0]
    # Save initial population
    current_N = N_0
    # Iterate for the specified time (t=0)
    for _ in range(time):
        # Update population using exponential growth formula
        current_N = current_N + p * current_N
        # Exponential growth formula
        current_N = round(current_N, 3)
        # Round to 3 decimal places
        N_values.append(current_N)
        # Append new population to the list
    return N_values
# Return the list of population values over time
```

```python
#Run the function with 40 days and p = 0.4 with an initial population of 0.01
result = N_exponential(0.01, 0.4, 35)
print(result)
#Display the results of the exponential growth equation
```

```python
# Plot the results with days on the x-axis and population on the y-axis
days = list(range(36))  # 0 to 35 days
plt.plot(days, result, marker='o')   
# Display the results of the exponential growth equation on a plot
plt.xlabel('Days')
# Label x-axis 'Days'
plt.ylabel(' Duckweed Population')
# Label y-axis 'Duckweed Population'
plt.title('Duckweed Population Growth Over Time')
# Title the plot 'Duckweed Population Growth Over Time'
plt.grid(True)  
# Add grid lines to the plot
plt.show()
# Show the plot
```
