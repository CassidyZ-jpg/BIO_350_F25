```python
# create a random forest model to examine the variable importance of latitude and longitiude on mammal alien species richness
# random forest model will be created using SciKit Learn package
# this random forest model will be used for a classification task
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
#load the cleaned data
data = pd.read_csv('Dawson_Cleaned.csv')
#select features and target variable
X = data[['LAT', 'LON']] #features: Latitude and Longitude
y = data['Mammals'] #target variable: Mammal alien species richness
print(np.shape(X)) #printing the shape of the feature set
print(np.shape(y)) #printing the shape of the target set
# split data set to train/test 20/80 split
from sklearn.model_selection import train_test_split #importing train_test_split function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #splitting the data into training and testing sets with 20% for testing
```

```python
#create the random forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42) #creating the random forest model with 100 trees
rf_model.fit(X_train, y_train) #fitting the model to the training data
#make predictions on the test set
y_pred = rf_model.predict(X_test) #making predictions on the test set
#evaluate the model
mse = mean_squared_error(y_test, y_pred) #calculating mean squared error
r2 = r2_score(y_test, y_pred) #calculating R^2 score
print(f'Mean Squared Error: {mse}') #printing mean squared error
print(f'R^2 Score: {r2}') #printing R^2 score
#examine variable importance
importances = rf_model.feature_importances_ #getting feature importances
feature_names = X.columns #getting feature names
for name, importance in zip(feature_names, importances):
    print(f'Feature: {name}, Importance: {importance}') #printing feature importances
```

```python
#visualize variable importance
plt.bar(feature_names, importances) #creating a bar plot of feature importances
plt.xlabel('Factors') #setting x-axis label
plt.ylabel('Variable Importance') #setting y-axis label
plt.title('Variable Importances of Lat and Long on Mammal Alien Species Richness') #setting title
plt.xticks(rotation=90) #rotating x-axis labels for better readability
plt.show() #displaying the plot
```

```python
# create a random forest model to examine the variable importance of latitude and longitiude on amphibian alien species richness
# random forest model will be created using SciKit Learn package
# this random forest model will be used for a classification task
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
#load the cleaned data
data = pd.read_csv('Dawson_Cleaned.csv')
#select features and target variable
X = data[['LAT', 'LON']] #features: Latitude and Longitude
y = data['Amphibians'] #target variable: Amphibian alien species richness
print(np.shape(X)) #printing the shape of the feature set
print(np.shape(y)) #printing the shape of the target set
# split data set to train/test 20/80 split
from sklearn.model_selection import train_test_split #importing train_test_split function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #splitting the data into training and testing sets with 20% for testing
```

```python
#create the random forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42) #creating the random forest model with 100 trees
rf_model.fit(X_train, y_train) #fitting the model to the training data
#make predictions on the test set
y_pred = rf_model.predict(X_test) #making predictions on the test set
#evaluate the model
mse = mean_squared_error(y_test, y_pred) #calculating mean squared error
r2 = r2_score(y_test, y_pred) #calculating R^2 score
print(f'Mean Squared Error: {mse}') #printing mean squared error
print(f'R^2 Score: {r2}') #printing R^2 score
#examine variable importance
importances = rf_model.feature_importances_ #getting feature importances
feature_names = X.columns #getting feature names
for name, importance in zip(feature_names, importances):
    print(f'Feature: {name}, Importance: {importance}') #printing feature importances
```

```python
#visualize variable importance
plt.bar(feature_names, importances) #creating a bar plot of feature importances
plt.xlabel('Factors') #setting x-axis label
plt.ylabel('Variable Importance') #setting y-axis label
plt.title('Variable Importances of Lat and Long on Amphibian Alien Species Richness') #setting title
plt.xticks(rotation=90) #rotating x-axis labels for better readability
plt.show() #displaying the plot
```
